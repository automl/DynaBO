```python
df = load_df()
```

```python
relevant_df = df[~df["end_date"].isnull()]
relevant_df = relevant_df[relevant_df["status"] == "done"]
relevant_df.head()
```

```python
# Sort by the amount of [ in incumbent_trace
relevant_df["new_incumbents"] = relevant_df["incumbent_trace"].apply(lambda x: x.count("["))
relevant_df = relevant_df.sort_values(by="new_incumbents", ascending=False)
```
