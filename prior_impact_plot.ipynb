```python
import numpy as np
import matplotlib.pyplot as plt

trials = 200
T = np.arange(start=0, stop=201, step=1)
prior_t = [50.0,100.0,150.0]

base = 0.5
exponent = lambda t, t_0 : (trials/10.0) / np.maximum((t - t_0)**2, 1E-8)

prior_traces = []
for prior in prior_t:
    prior_x = np.arange(prior, 201, 1)
    prior_y = exponent(prior_x, prior)
    prior_traces.append((prior_x, prior_y))


plt.figure(figsize=(6,5))
i = 1
for trace in prior_traces:
    plt.plot(trace[0], np.array([1 - np.power(base, y_exp) for y_exp in trace[1]]), label="$\pi^{("+str(i)+")}$")
    i += 1
        
plt.rc('font', size=12)
plt.rc('axes', titlesize=20)     # fontsize of the axes title
plt.rc('axes', labelsize=30)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
plt.rc('legend', fontsize=20)            # legend fontsize
#plt.title(f'Prior Impact over Iterations for lcbench and rbv2_xgboost')
plt.xlabel('Iteration')
plt.ylabel('Prior Impact')
plt.xlim(left=0)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('PriorImpact-yahpo.pdf')
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

trials = 50

T = np.arange(start=0, stop=trials+1, step=1)
prior_t = [10.0,20.0,30.0]

base = 0.5
exponent = lambda t, t_0 : (trials/10) / np.maximum((t - t_0), 1E-8)

prior_traces = []
for prior in prior_t:
    prior_x = np.arange(prior, trials+1, 1)
    prior_y = exponent(prior_x, prior)
    prior_traces.append((prior_x, prior_y))


plt.figure(figsize=(20,7))
sns.set(style='whitegrid', font_scale=4)
i = 1
for trace in prior_traces:
    sns.lineplot(x=trace[0], y=np.array([1 - np.power(base, y_exp) for y_exp in trace[1]]), label="$\pi^{("+str(i)+")}$", linewidth=5)
    i += 1
plt.xlabel('Number of Evaluations', fontweight='bold')
plt.ylabel('Prior Impact', fontweight='bold')
plt.savefig('PriorImpact-50.pdf', bbox_inches="tight")
plt.show()
plt.close()
# SMALL_SIZE=20
# plt.rc('font', size=22)
# plt.rc('axes', titlesize=30, labelsize=30)     # fontsize of the axes title
# plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
# plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
#
# # plt.title(f'Prior Impact over Iterations for PD1 Benchmarks')
# plt.xlim(left=0)
# plt.legend(
#     loc="upper center",
#     bbox_to_anchor=(0.5, -0.3),
#     ncol=3,
#     fontsize=30,
# )
#
# plt.grid(True)
# plt.tight_layout()
# plt.savefig('PriorImpact-50.pdf')
# plt.show()
```
